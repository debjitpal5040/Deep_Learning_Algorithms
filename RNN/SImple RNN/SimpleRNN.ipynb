```python
import numpy as np
from keras.models import Sequential
from keras.layers import SimpleRNN, Dense, Input, LSTM

# Preparing data
X = np.array([1, 4, 9, 16, 25, 36, 49, 64])
y = np.array([4, 9, 16, 25, 36, 49, 64, 81])

# Reshaping data for the model
X = X.reshape((1, len(X), 1))
y = y.reshape((1, len(y), 1))

# Defining the model
model = Sequential()
model.add(Input((8, 1)))
model.add(LSTM(500, activation="tanh"))
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse")

# Training the model
model.fit(X, y, epochs=200, verbose=1)

# Predicting the next number
x_input = np.array([4, 9, 16, 25, 36, 49, 64, 81])
x_input = x_input.reshape((1, len(x_input), 1))
yhat = model.predict(x_input, verbose=1)[0]

print(f"The predicted next number in the sequence is: {yhat}")
```
